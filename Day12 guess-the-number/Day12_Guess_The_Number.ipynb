```python
import random

def number_Generator():
    return random.randint(1,100)

def choosing_Attempts():
    print('Choose the level...')
    level=input("'h' for hard\n'm' for medium\n'e' for easy\n:")
    if level=='h':
        return 6
    elif level=='m':
        return 8
    elif level=='e':
        return 10
    else:
        print('PLEASE, Enter the level as given in criteria')
        choosing_Attempts()

def finding_number(attempts,actualNumber):
    totalAttempts = attempts
    while(attempts):
        print(f'You got {attempts} number of attempts!!!')
        guessedNumber = int(input('Guess the number : '))

        if guessedNumber== actualNumber:
            print(f"You got the right number: '{actualNumber}' in {totalAttempts-attempts+1} Attempts")
            return
        elif guessedNumber < actualNumber:
            print('Guess is too low..\nGo for the high number!')
        else:
            print('Guess is too high..\nGo for the low number!')

        attempts-=1
    print("\nBetter luck next time!!!")
    print(f"Actual number was: {actualNumber}")

end_of_the_game=False
while not end_of_the_game:
    print('WELCOME TO THE GAME: GUESS THE NUMBER ')

    attempts = choosing_Attempts()
    number = number_Generator()
    finding_number(attempts,number)

    cont=input(f"Do you wants to continue the game (Type 'y' for yes and 'n' for no) : ").lower()
    if cont=='y':
        end_of_the_game=False
    else:
        end_of_the_game=True
        print('Thank you for playing the game!!!')
```
