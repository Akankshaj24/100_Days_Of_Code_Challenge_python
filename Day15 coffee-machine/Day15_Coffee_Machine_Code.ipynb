```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}
profit = 0


def printing_report():
    print("Current resource values :")
    print(f"Water : {resources['water']}ml")
    print(f"Milk : {resources['milk']}ml")
    print(f"Coffee : {resources['coffee']}g")
    print(f"Money : {profit}$")


def is_sufficient_resources(ingredient):
    """Checks for the sufficient resources"""
    for item in ingredient:
        if ingredient[item] > resources[item]:
            return False
    return True


def money_inserted():
    """Returns the Amount inserted by user"""
    total = 0
    total += int(input("How many Quarters : ")) * 0.25
    total += int(input("How many Dimes : ")) * 0.10
    total += int(input("How many nickles : ")) * 0.05
    total += int(input("How many Pennies : ")) * 0.01
    return total


def is_transaction_successful(money, actual_cost):
    if money >= actual_cost:
        global profit
        profit += actual_cost
        change = money - actual_cost
        print(f"Here is ${round(change, 2)} dollars in change.")
        return True
    else:
        print("Sorry, that's not enough money!")
        return False


def resource_update(ingredient):
    for item in ingredient:
        resources[item] -= ingredient[item]


print('WELCOME TO COFFEE MACHINE PROGRAM !!')
is_on = True
while is_on:
    choice = input("“What would you like? (espresso/latte/cappuccino):”")
    if choice == 'off':
        is_on = False
    elif choice == 'report':
        printing_report()
    else:
        drink = MENU[choice]
        if is_sufficient_resources(drink["ingredients"]):
            inserted_money = money_inserted()
            if is_transaction_successful(inserted_money, drink["cost"]):
                resource_update(drink["ingredients"])
                print(f"Here is your {choice}, Enjoy!!!")
```
